```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver=webdriver.Chrome(r"C:\Users\DELL\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
url='https://www.billboard.com/'
driver.get(url)
```

```python
search_list=driver.find_element_by_id('standalone-header-search-keys')
search_list
```

```python
search_list=driver.find_element_by_xpath('//a[@class="header__main-link header__main-link--charts"]')
search_list
```

```python
Hot_list=driver.find_element_by_xpath('//a[@class="header__submenu__list__link"]')
Hot_list
```

```python
song_name=driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']")
song_name
```

```python
song_list=[]
for i in song_name:
    song_list.append(i.text)
song_list    
```

```python
Artist_name=driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']")
Artist_name
```

```python
Artist_list=[]
for i in Artist_name:
    Artist_list.append(i.text)
Artist_list    
```

```python
week_rank=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
week_rank
```

```python
week_list=[]
for i in week_rank:
    week_list.append(i.text)
week_list    
```

```python
peak_rank=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
peak_rank
```

```python
peak_list=[]
for i in peak_rank:
    peak_list.append(i.text)
peak_list  
```

```python
Weekon_board=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
Weekon_board
```

```python
weekon_rank=[]
for i in Weekon_board:
    weekon_rank.append(i.text)
weekon_rank
```

```python
#creating dataframe
df=pd.DataFrame({})
df['song']=song_list
df['Artist']=Artist_list
df['week rank']=week_list
df['peak rank']=peak_list
df['week on board']=weekon_rank
```

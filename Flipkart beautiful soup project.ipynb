```python
from bs4 import BeautifulSoup as soup
import requests
import csv
import pandas as pd
```

```python
url='https://www.flipkart.com/search?q=pixel+4a&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&as-pos=1&as-type=RECENT&suggestionId=pixel+4a%7CMobiles&requestId=9815a74f-56a8-432b-ad4c-bea5cdd52434&as-searchtext=pixel'
response = requests.get(url)
htmlcontent = response.content
soup = BeautifulSoup(htmlcontent,"html.parser")
print(soup.prettify)
```

```python
product=[]
price=[]
rating=[]
battery=[]
Display=[]
RAM=[]
Camera=[]
Processor=[]
product=soup.find('div',attrs={'class':'_4rR01T'})
print(product.text)
```

```python
for a in soup.findAll('a',href=True, attrs={'class':'_2kHMtA'}):
    name=a.find('div',attrs={'class':'_4rR01T'})
    price=a.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
    rating=a.find('div',attrs={'class':'_3LWZlK'})
    battery=a.find('div',attrs={'class':'rgWa7D'})
    Display=a.find('div',attrs={'class':'rgWa7D'})
    RAM=a.find('div',attrs={'class':'rgWa7D'})
    Camera=a.find('div',attrs={'class':'rgWa7D'})
    Processor=a.find('div',attrs={'class':'rgWa7D'})
    products.append(name.text)
    price.append(price.text)
    rating.append(rating.text)               
    battery.append(battery.text)
    Display.append(Display.text)
    RAM.append(RAM.text)
    Camera.append(Camera.text)
    Processor.append(Processor.text)
```

```python
df = pd.DataFrame({'Product Name':product,'Price':price,'Rating':rating,'battery':battery,'Display':Display,'RAM':RAM,'Camera':Camera,'Processor':Processor})
df.head()
```

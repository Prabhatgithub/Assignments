```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
url='https://www.imdb.com/list/ls095964455/'
driver.get(url)
```

```python
search_list=driver.find_elements_by_xpath("//input[@class='imdb-header-search__input _3gDVKsXm3b_VAMhhSw1haV react-autosuggest__input']")
search_list
```

```python
series_name=driver.find_elements_by_xpath("//h3[@class='lister-item-header']")
series_name
```

```python
series_list=[]
for i in series_name:
    series_list.append(i.text)
series_list 
```

```python
year_span=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
year_span
```

```python
year_list=[]
for i in year_span:
    year_list.append(i.text)
year_list 
```

```python
genre_name=driver.find_elements_by_xpath("//span[@class='genre']")
genre_name
```

```python
genre_list=[]
for i in genre_name:
    genre_list.append(i.text)
genre_list 
```

```python
Run_time=driver.find_elements_by_xpath("//span[@class='runtime']")
Run_time
```

```python
run_list=[]
for i in Run_time:
    run_list.append(i.text)
run_list 
```

```python
Rating_stars=driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']//span")
Rating_stars
```

```python
rating_list=[]
for i in Rating_stars:
    rating_list.append(i.text)
rating_list 
```

```python
Rating = []

for i in range(1,len(rating_list),2):
    Rating.append(rating_list[i])
    
Rating
```

```python
Series_votes=driver.find_elements_by_xpath("//span[@name='nv']")
Series_votes
```

```python
votes_list=[]
for i in Series_votes:
    votes_list.append(i.text)
votes_list 
```

```python
#creating dataframe
df=pd.DataFrame({})
df['name']=series_list
df['year']=year_list
df['genre']=genre_list
df['time']=run_list
df['rating']=Rating
df['votes']=votes_list
df
```
